### Исправление после ревью.

Найдем событие с наибольшим разбросом интервалов времени до его наступления в рамках одной сессии.

```python
def summary(x):                                           # создадим функцию для нахождения min и max значения 'time_differens'
    result = {                                            # и их разницы
        'time_differens_min': x['time_differens'].min(),
        'time_differens_max': x['time_differens'].max(),
        'interval': x['time_differens'].max() - x['time_differens'].min()
    }
    return pd.Series(result)
```

```python
max_interval_for_event = (                                # применим функцию к данным с группировкой по 'event_id' и 'sender'
    events.
    dropna(subset=['time_differens']).groupby(['event_id', 'sender']).
    apply(summary).
    reset_index()
)
max_interval_for_event
```

```python
max_interval_for_event[max_interval_for_event['interval']==max_interval_for_event['interval'].max()]     # найдем max интервал
```

### Вопрос 4.
Какое количество задач прогнозирования могут выполняться одновременно? Укажите максимальное число параллельных задач.

```python
forecast_marker_duration_31 = forecast_marker_duration_31.sort_values(by='d')          # отсортируем по дате запуска задачи
```

```python
def maximumOverlap(tasks):                            # создадим функцию для подсчета максимального числа параллельных задач
    times = []
    for task in tasks:
        startTime, endTime = task
        times.append((startTime, 'start'))
        times.append((endTime, 'end'))
    times = sorted(times)

    count = 0
    maxCount = 0
    for time in times:
        if time[1] == 'start':
            count += 1    # increment on arrival/start
        else:
            count -= 1    # decrement on departure/end
        maxCount = max(count, maxCount)  # maintain maximum
    return maxCount

zipped = zip(forecast_marker_duration_31['d'].to_list(), forecast_marker_duration_31['request_time'].to_list())
zipped                                       # передадим в функцию диапазоны с датой и временем запуска задач и их завершения
tasks = zipped
print('Максимальное число параллельных задач равно', maximumOverlap(tasks))
```

### Ответ:

В определенный промежуток времени 5 задач прогнозирования выполняются одновременно.
